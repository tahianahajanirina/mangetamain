```python
from transformers import pipeline

# Charger votre modèle depuis Hugging Face
sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="TahianaAndriambahoaka/sentiment-analysis-food-reviews"
)

def predict_sentiment(text):
    """Prédit le sentiment d'un texte"""

    result = sentiment_analyzer(text)
    print(result)

    return result[0]['label'], result[0]['score']


# Analyser plusieurs commentaires en une fois
reviews = [
    "Amazing! Will make again.",
    "Too salty, not recommended.",
    "It was okay, nothing special.",
    "good"
]
for review in reviews:
    label, score = predict_sentiment(review)
    print(f"Review: {review}\nSentiment: {label}, Confidence: {score}\n")
```
